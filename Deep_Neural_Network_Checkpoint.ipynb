##Cleaning our dataset

```python
bank.nunique()
```

```python
bank.drop(columns=['age'	,'job'	,'marital'	,'education','day'	,'month'],inplace=True)
bank
```

```python
bank.info()
```

```python
# Convert categorical variables to numerical
from sklearn.preprocessing import LabelEncoder 

bank["default"]=bank["default"].map({"yes": 1, "no": 0}) 
bank["housing"]=bank["housing"].map({"yes": 1, "no": 0}) 
bank["loan"]=bank["loan"].map({"yes": 1, "no": 0}) 
bank["deposit"]=bank["deposit"].map({"yes": 1, "no": 0}) 

bank["contact"]=LabelEncoder().fit_transform(bank["contact"])
bank["poutcome"]=LabelEncoder().fit_transform(bank["poutcome"])
```

##Building a deep learning network

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

# features and target
X = bank.drop('deposit', axis=1)
y = bank['deposit']
```

```python
#spliting our data
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.1)
X_model, X_valid, y_model, y_valid = train_test_split(X_train, y_train, test_size = 0.2)
```

```python
#define our model
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Training our model
model.fit(X_model, y_model, validation_data=(X_valid, y_valid), epochs= 20) 
```

```python
# Evaluate our model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test accuracy: {accuracy}')
```
