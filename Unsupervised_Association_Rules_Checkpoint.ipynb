```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
#Importing libraries:
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori 
from mlxtend.frequent_patterns import association_rules 
from mlxtend.frequent_patterns import association_rules 
```

```python
#Apriori Preparation:
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)  
df=pd.DataFrame(te_ary, columns=te.columns_) 
df
```

```python
#Support metric: "select itemsets with a minimum of 60% Support"
apriori(df, min_support=0.6)
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
#Confidence metric:
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) 
```

```python
#Lift metric:
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

#Other dataset:

```python
data=pd.read_csv("Market_Basket_Optimisation.csv")
data
```

```python
data.info()
```

```python
transactions=[]
for i in range(0,7500):
    transactions.append([str(data.values[i,j]) for j in range (0,20) ])
transactions
```

```python
#Apriori Preparation:
te=TransactionEncoder()
te_ary=te.fit(transactions).transform(transactions)  
df2=pd.DataFrame(te_ary, columns=te.columns_) 
df2
```

```python
#Support metric: "select itemsets with a minimum of 1% Support"
apriori(df2, min_support=0.01)
frequent_itemsets=apriori(df2, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
#Confidence metric:
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.1)
```

```python
#Lift metric:
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
